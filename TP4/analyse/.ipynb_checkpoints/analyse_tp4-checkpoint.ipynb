# Importation des librairies

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, probplot, shapiro, spearmanr
```

# lecture des donnée

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.rename(columns={'commit_id':'commit', 'nc_list':'NC', 'mBC':'mcBC'}, inplace = True)
```

```python
df = df.iloc[::-1] #reverse commit order
```

```python
df['index'] = list(range(0,len(df)))
```

```python
df.head()
```

# Tâche 4

Montrez visuellement l’évolution des métriques NC, mWMC, et mcBC à travers l’histoire de jfreechart et essayez
d’expliquer de façon qualitative ce que vous observez.

```python
px.line(df,x='index', y='NC', hover_data=['commit'])
```

```python
px.line(df,x='index', y='mWMC', hover_data=['commit'])
```

```python
px.line(df,x='index', y='mcBC', hover_data=['commit'])
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 10), sharex=True)
yLabels = ['nombre de classes', 'moyenne de WMC', 'moyenne de BC']
plt.suptitle('Métriques du commit le plus ancien au plus récent')
ax1.plot(df['index'].values, df.NC.values, label='NC', color='b')
ax2.plot(df['index'].values, df.mWMC.values, label='mWMC', color='k')
ax3.plot(df['index'].values, df.mcBC.values, label='mcBC', color='r')
for ax in [ax1,ax2,ax3]:
    ax.legend(loc='best')
    
for i,ax in enumerate([ax1,ax2,ax3]):
    ax.set_ylabel(yLabels[i])
plt.tight_layout()
plt.savefig('tache4.png', dpi=500)
```

Évaluez l’hypothèse selon laquelle pour les versions du jfreechart il existe une **corrélation** entre les métriques **NC**
et **mWMC** et entre les métriques **NC** et **mcBC**. Basez votre analyse sur l’hypothèse que leurs valeurs **ne sont pas
normalement distribuées**.

```python
metrics = df.loc[:, ['NC', 'mWMC', 'mcBC']]
```

```python
sns.pairplot(metrics, diag_kind='kde')
```

### Test de statistique pour la corrélation

Tel qu'énoncé, on fait l'hypothèse que les données ne sont pas distribué normalement. Donc, nous allons utilisé le test de spearman

### NC vs mWMC

```python
spearmanr(df.NC.values, df.mWMC.values, alternative='two-sided')
```
